```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
data = {
    "도시": ["안양", "안양", "안양", "금천구", "금천구", "금천구", "수원", "수원", "수원"],
    "연도": ["2012", "2016", "2020", "2012", "2016", "2020", "2012", "2016", "2020"],
    "인구": [611400, 597400, 550000, 243300, 235400, 231700, 1120000, 1194000, 1186000],
    "지역": ["경기도", "경기도", "경기도", "서울", "서울", "서울", "경기도", "경기도", "경기도"]
}
columns = ["도시", "연도", "인구", "지역"]
pop = pd.DataFrame(data, columns=columns)
pop
```

```python
pop.pivot("연도","도시","인구") # 도시별 인구 증감폭
```

```python
pop.set_index(["연도", "도시"])[["인구"]].unstack() # 동일데이터
```

```python
pop.pivot(["지역", "도시"], "연도", "인구") # 경기도/ 서울 분할
```

```python
korea_groups = pop.groupby(pop.연도) # group 데이터 생성
korea_groups
```

```python
korea_groups.groups
```

```python
pop.인구.groupby(pop.연도).sum() # 연도별 안양+금천구+수원시 합계 시리즈로 호출
```

```python
korea_groups.sum() # 연도별 안양+금천구+수원시 합계 데이터프레임으로 호출
```
